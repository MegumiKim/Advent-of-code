```python
#Part 1

with open ("input") as f:
     passport_list = f.read().split("\n\n")

passports = [passport.split() for passport in passport_list]
print(passports[0])
passports_dict = [make_passport_dict(passport) for passport in passports]


def make_passport_dict(l):
    return {field.split(":")[0]:field.split(":")[1] for field in l}


def is_valid_passport(passport):
    fields = {"byr", "iyr", "eyr", "hgt", "hcl", "ecl", "pid"}
    return fields.issubset(passport.keys())

                        
def count_valid_passports (passports):
    valid_passports = 0
    
    for passport in passports:
        if is_valid_passport(passport):
            valid_passports += 1
    return valid_passports
  

print("Part1:", count_valid_passports(passports_dict))
    
```

```python
#Part 2

def is_valid_passport_part2(passport):

    def valid_years (bry,iyr,eyr):
        try:
            if 1920 <= int(bry) <= 2002 and \
               2010 <= int(iyr) <= 2020 and \
               2020 <= int(eyr) <= 2030:
                return True
        except ValueError:
            return False
        return False

    
    def valid_hgt(hgt):
        unit = hgt[-2:]

        try:
            hgt = int(hgt.replace(unit,""))
        except ValueError:
            return False

        if unit == "cm":
            return 150 <= hgt <=193
        elif unit == "in":
            return 59 <= hgt <=76
        return False


    def valid_hcl(hcl):
        if hcl[0] != "#" or len(hcl) != 7:
            return False    
        for i in hcl[1:]:
            if i not in "0123456789abcdef": 
                return False
        return True


    def valid_ecl(ecl):
        valid_color = ["amb", "blu", "brn", "gry", "grn", "hzl", "oth"]
        if ecl in valid_color:
            return True
        return False


    def valid_pid(pid):
        return len(pid) == 9 and pid.isdecimal()

    
    fields = {"byr", "iyr", "eyr", "hgt", "hcl", "ecl", "pid"}

    if fields.issubset(passport.keys())\
    and valid_years(passport["byr"],passport["iyr"], passport["eyr"])\
    and valid_hgt(passport["hgt"])\
    and valid_hcl(passport["hcl"])\
    and valid_ecl(passport["ecl"])\
    and valid_pid(passport["pid"]):       
        return True
    return False


def count_valid_passports_part2 (passports):
    valid_passports = 0
    for passport in passports:
        if is_valid_passport_part2(passport):
            valid_passports += 1
    return valid_passports


print("Part 2:", count_valid_passports_part2(passports_dict))
```
