```python
with open ('input') as file:
    inputs = file.read().splitlines()
```

```python
# Making dictionaries within dictionaries

bags_dict = {}

for line in inputs:
    line = line.replace('bags', 'bag').replace('bag.', 'bag').replace(' bag', '')
    key, value = line.split(" contain ")
    
    if "no other" in value:
        bags_dict[key] = {}
   
    else:
        bags_dict[key] = value

        sub_bags = {}
        for sub in value.split(', '):
            sub_bags[sub[2:]] = sub[0]    
               
        bags_dict[key] = sub_bags

bags_dict
```

```python
#Part 1: How many bag colors can eventually contain at least one shiny gold bag? 

def bags_counter(my_bag, dictionary):
    outer_bags = []

    for index in dictionary:
        if my_bag in dictionary[index]:
            outer_bags.append(index)
            outer_bags.extend(bags_counter(index, dictionary))
    return set(outer_bags)

print('Part1:', len(bags_counter('shiny gold', bags_dict)))
```

```python
#Part 2: How many individual bags are required inside your single shiny gold bag?

def bags_counter_part2(my_bag, dictionary):
    bags_counter = 1   
    inner_bags = dictionary[my_bag]
    
    for item in inner_bags:
        count = int(inner_bags[item])
        bags_counter += count*bags_counter_part2(item, dictionary)        
    return bags_counter
    
print('Part 2:', bags_counter_part2('shiny gold', bags_dict)-1)
```
